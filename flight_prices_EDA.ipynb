```python
import re
import datetime

def convert_travel_time(time_string):
    match = re.match(r"P(\d+)?D?T?(\d+)?H?(\d+)?M?", time_string)
    if not match:
        raise ValueError("Invalid travel time format: {}".format(time_string))

    d = match.group(1)
    h = match.group(2)
    m = match.group(3)

    if d == None:
      d= 0
    if h == None:
      h = 0
    if m == None:
      m = 0

    days = int(d)
    hours = int(h)
    minutes = int(m)

    # Create a datetime.timedelta object representing the travel time
    travel_time = datetime.timedelta(days= days, hours=hours, minutes=minutes)

    return travel_time
```

```python
df1['travelDuration'] = df1['travelDuration'].apply(convert_travel_time)
```
